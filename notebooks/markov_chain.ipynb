# Markov Chain - simple example #

Let's consider a simple example of a Markov chain in neuroscience: the Hodgkin-Huxley model of action potential generation in neurons. This model describes how the opening and closing of ion channels in the neuron's membrane can lead to the generation of an action potential.

In the Hodgkin-Huxley model, the state of an ion channel (open or closed) is a Markov process. The probability of transitioning from the open state to the closed state, or vice versa, depends only on the current state of the channel and not on how it got to that state.

Here's a simplified version of this model:

1. **State 1 - Closed**: The ion channel is closed. The probability of transitioning to the open state depends on the membrane potential.

2. **State 2 - Open**: The ion channel is open, allowing ions to flow through and changing the membrane potential. The probability of transitioning back to the closed state also depends on the membrane potential.

This can be represented as a transition matrix:



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Transition matrix
transition_matrix = np.array([
    [0.9, 0.1],  # Probabilities of staying in the closed state and transitioning to the open state
    [0.1, 0.9]   # Probabilities of transitioning to the closed state and staying in the open state
])
```

In this transition matrix, `transition_matrix[i, j]` is the probability of transitioning from state i to state j. For example, `transition_matrix[0, 1]` is the probability of transitioning from the closed state (state 0) to the open state (state 1).

To simulate this Markov chain, you would start in one of the states, then use the transition probabilities to determine the state at the next time step, and repeat this process for as many time steps as you want to simulate.

```python
# Define the transition matrix
transition_matrix = np.array([
    [0.9, 0.1],  # Probabilities of staying in the closed state and transitioning to the open state
    [0.1, 0.9]   # Probabilities of transitioning to the closed state and staying in the open state
])

# Define the initial states
initial_states = [0.2, 0.5]
# initial_states = np.linspace(0, 1, 11)

# Define the number of time steps
num_time_steps = 100

# Initialize an array to store the state at each time step
states = np.zeros((len(initial_states), num_time_steps), dtype=int)

# Simulate the Markov chain for each initial state
for i, initial_state in enumerate(initial_states):
    # Set the initial state
    states[i, 0] = initial_state

    # Simulate the Markov chain
    for t in range(1, num_time_steps):
        # Determine the next state
        states[i, t] = np.random.choice([0, 1], p=transition_matrix[states[i, t-1]])

# Plot the state at each time step for each initial state
for i, initial_state in enumerate(initial_states):
    plt.plot(states[i], label=f'Initial state: {initial_state}')

plt.xlabel('Time step')
plt.ylabel('State')
plt.title('Markov chain for different initial states')
plt.legend()
plt.show()
```
